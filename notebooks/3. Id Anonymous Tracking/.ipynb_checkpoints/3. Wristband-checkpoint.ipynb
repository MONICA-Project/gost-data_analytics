## Functions

```python
def extract_array_json(cursor):
    """This function reads the cluster, extract the json files, and transfer everything 
    in a dictionnary with the timestamp as key and the heat map as value. 
    Time_HM is a list of all keys sorted. We also export the dictionnary and list with 
    datastamp under a string format for easier display later"""
    
    Dico_latlon = {}
    Dico_latlon_cumul = {}
    Time_latlon = []
    lat_all = []
    lon_all = []
    for row in cursor:
        json_data = str(row[1])
        b = ast.literal_eval(json_data)

        #We check data to get latitude and longitude
        lat = b['result']["lat"]
        lon = b['result']["lon"]

        #We get the timestamp
        Time = dateutil.parser.parse(b["phenomenonTime"])
        Time = Time + timedelta(hours=2)
        
        #We fill the lists and dictionaries
        Time_latlon.append(Time)
        lat_all.append(lat)
        lon_all.append(lon)
        a= list(Dico_latlon.values())
        b = a + [(lat, lon)]
        Dico_latlon[Time] = (lat, lon)
        Dico_latlon_cumul[Time] = b
    Time_latlon=sorted(Time_latlon)
    return Dico_latlon,Dico_latlon_cumul, Time_latlon, lat_all, lon_all
```

```python
def Extract_loc_Hr (hr, Dico):
    """function that returns the value (location) for a key (timestamp)"""
    loc = (Dico[hr])
    return loc
```
